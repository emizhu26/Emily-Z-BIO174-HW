__Exercise 2.__ Modify the query above to return the gene name in yeast, the gene names in the other three species, the cai value and the sequence for all yeat genes with a cai value larger that 0.8 and that have orthologs in the other three species.

```python
pd.read_sql("""SELECT s1.gene_name AS gene_name,
  o.mikatae AS mikatae,
  o.paradoxus AS paradoxus,
  o.bayanus AS bayanus,
  c.CAI AS cai,
  s1.seq AS yeast_seq
FROM sequences AS s1
JOIN orthologs AS o ON s1.gene_name = o.cerevisiae
JOIN cai AS c ON s1.gene_name = c.title
WHERE c.CAI > 0.8
  AND o.mikatae IS NOT NULL
  AND o.paradoxus IS NOT NULL
  AND o.bayanus IS NOT NULL""", con)
```

__Exercise 3.__ Perform the same query as above, but this time include the chromosome number of the gene in the output.

```python
pd.read_sql("""SELECT s1.gene_name AS gene_name,
       o.mikatae AS mikatae,
       o.paradoxus AS paradoxus,
       o.bayanus AS bayanus,
       c.CAI AS cai,
       s1.seq AS yeast_seq,
       g.chrom AS chrom_number
FROM sequences AS s1
JOIN orthologs AS o ON s1.gene_name = o.cerevisiae
JOIN cai AS c ON s1.gene_name = c.title
JOIN gff AS g ON s1.gene_name = g.acc
WHERE c.CAI > 0.8
  AND o.mikatae IS NOT NULL
  AND o.paradoxus IS NOT NULL
  AND o.bayanus IS NOT NULL""", con)
```

__Exercise 4.__ Now generate a table with the number of genes per chromosome that have CAI larger than 0.5 and have orthologs in the other three yeast species.

```python
pd.read_sql("""SELECT g.chrom AS chromosome,
       COUNT(s1.gene_name) AS num_genes
FROM sequences AS s1
JOIN orthologs AS o ON s1.gene_name = o.cerevisiae
JOIN cai AS c ON s1.gene_name = c.title
JOIN gff AS g ON s1.gene_name = g.acc
WHERE c.CAI > 0.5
  AND o.mikatae IS NOT NULL
  AND o.paradoxus IS NOT NULL
  AND o.bayanus IS NOT NULL
GROUP BY g.chrom""", con)
```
